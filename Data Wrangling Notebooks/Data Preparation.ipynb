# Creating a Season Variable (Categorical Value)

Winter: December, January, February

Spring: March, April, May

Summer: June, July, August

Fall: September, October, November

Format in the data: dd/mm/yyyy

```python
df["Date"]
```

```python
for index, value in df["Date"].items():
    
    print(str(index) + ": " + str(value))
    #print(type(value))
    
    print("Day: " + value[0] + value[1])
    print("Month: " + value[3] + value[4])
    print("Year: 20" + value[-2] + value[-1])
    print()
```

```python
for index, value in df["Date"].iteritems():
    
    # m1 m2 → mm format
    m1 = value[3]
    m2 = value[4]

    print("\n" + str(index))
    
    # Winter -- 12, 01, or 02
    if (m1 == '1' and m2 == '2') or (m1 == '0' and m2 == '1') or(m1 == '0' and m2 == '2'):
        print(m1+m2 + " → Winter")
        df.at[index, "Season"] = "Winter"
        
    # Spring -- 03, 04, or 05
    if (m1 == '0' and m2 == '3') or (m1 == '0' and m2 == '4') or(m1 == '0' and m2 == '5'):
        print(m1+m2 + " → Spring")
        df.at[index, "Season"] = "Spring"
        
    # Summer -- 06, 07, or 08
    if (m1 == '0' and m2 == '6') or (m1 == '0' and m2 == '7') or(m1 == '0' and m2 == '8'):
        print(m1+m2 + " → Summer")
        df.at[index, "Season"] = "Summer"
        
    # Fall -- 09, 10, or 11
    if (m1 == '0' and m2 == '9') or (m1 == '1' and m2 == '0') or(m1 == '1' and m2 == '1'):
        print(m1+m2 + " → Fall")
        df.at[index, "Season"] = "Fall"
```

```python
def find_number_of_fouls(referee):
    total = 0
    
    for a in df.itertuples(index=False):
        if a[11] == referee:
            total += a[15]+ a[16]
        
    return total
```
